```python
import numpy as np
arr = np.array([[-1, 2, 0, 4], [4, -0.5, 6, 0], [2.6, 0, 7, 8], [3, -7, 4, 2.0]])
print("Original array:\n", arr)
```

```python
print("\nEvery other rows: \n", arr[0:3:2])
```

```python
arra = np.array([1, 2, 3])
arrb = np.array([4, 5, 6])
arr = np.concatenate((arra, arrb))
print("\n Original arrays:\n",arra,arrb) 
print("\n Joined array:\n",arr)
```

```python
#Horizondal join
arr= np.hstack((arra, arrb))
print("\n Horizondal joining:\n",arr)
```

```python
arr = np.vstack((arra, arrb))
print("\n Vertical joining:\n", arr)
```

```python
arr = np.dstack((arra, arrb))
print("\n Depth joining:\n",arr)
```

```python
arr= np.array([1, 2, 3, 4, 5, 6]) 
newarr= np.array_split(arr, 3)
print("\nOriginal Array:\n",arr)
print("\nSplitted array:\n",newarr)
print("\nSplitted array in another form:\n")
print(newarr[0])
print(newarr[1])
print(newarr[2])
```
